### github notes

key terms:
* **repository** - a container holding all *committed* files and folders
* **commit** - saving a file version in a repository (along with a save name, i.e. commit message)
* **git** - this is a version control program (i.e. a app for saving many versions of files, ex. computer code.)
* **github.com** - an online company (website) providing a place for users to save their respositories.
